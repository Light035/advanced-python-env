```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {
    "Name": ["Alex", "Maria", "John", "Sara", "Tom"],
    "Score": [75, 88, 92, 81, 69]
}

df = pd.DataFrame(data)
```

```python
plt.plot(df["Name"], df["Score"])
plt.xlabel("Students")
plt.ylabel("Scores")
plt.title("Student Scores Line Plot")
plt.show()
```

```python
plt.bar(df["Name"], df["Score"])
plt.xlabel("Students")
plt.ylabel("Scores")
plt.title("Student Scores Bar Chart")
plt.show()
```
